##Numerical Differentiation Problem Statement

How can we generate a numerical grid for evaluating a function over a domain?

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate a numerical grid
def generate_grid(x_start, x_end, y_start, y_end, num_points_x, num_points_y):
    x = np.linspace(x_start, x_end, num_points_x)
    y = np.linspace(y_start, y_end, num_points_y)
    X, Y = np.meshgrid(x, y)
    return X, Y

# Define grid parameters
x_start, x_end = -np.pi, np.pi
y_start, y_end = -np.pi, np.pi
num_points_x, num_points_y = 50, 50

# Generate grid
X, Y = generate_grid(x_start, x_end, y_start, y_end, num_points_x, num_points_y)

# Evaluate a function on the grid
Z = np.exp(-X**2 - Y**2)  # Example function

# Plot numerical grid
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='coolwarm')
plt.colorbar(label='Function Value')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Numerical Grid for Function Evaluation')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return np.exp(x)

# Define step size
h = 0.1
x = np.arange(0, 2, h)  # Define domain
y = f(x)  # Compute function values

# Compute forward difference approximation
forward_diff = (f(x + h) - f(x)) / h

# Compute exact derivative
exact_derivative = np.exp(x)

# Plot results
plt.figure(figsize=(8, 6))
plt.plot(x, forward_diff, 'r--', label='Forward Difference Approximation')
plt.plot(x, exact_derivative, 'b-', label='Exact Derivative')
plt.xlabel('x')
plt.ylabel('Derivative')
plt.legend()
plt.title('Forward Difference Approximation vs. Exact Derivative')
plt.show()

# Compute maximum error
max_error = np.max(np.abs(forward_diff - exact_derivative))
print(f"Max Error: {max_error}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return np.sin(x)

# Define step size
h = 0.1
x = np.arange(0, 2*np.pi, h)  # Define domain
y = f(x)

# Compute second derivative using central difference
second_derivative = (f(x + h) - 2*f(x) + f(x - h)) / h**2

# Compute exact second derivative
exact_second_derivative = -np.sin(x)

# Plot results
plt.figure(figsize=(8, 6))
plt.plot(x, second_derivative, 'r--', label='Numerical Approximation')
plt.plot(x, exact_second_derivative, 'b-', label='Exact Second Derivative')
plt.xlabel('x')
plt.ylabel('Second Derivative')
plt.legend()
plt.title('Second Derivative Approximation using Central Difference')
plt.show()

# Compute maximum error
max_error = np.max(np.abs(second_derivative - exact_second_derivative))
print(f"Max Error: {max_error}")
```
