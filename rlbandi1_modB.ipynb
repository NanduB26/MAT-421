##Interpolation Problem Statement

Interpolation involves determining a function that precisely passes through a given set of data points. Unlike regression, interpolation assumes the data is exact and seeks to fit the function through every point without deviation. Given the data: x=[0,2,4,6,8] and y=[1,4,2,5,7], create a plot showing the data points and explain why interpolation might be useful in this context.

```python
import matplotlib.pyplot as plt

x = [0, 2, 4, 6, 8]
y = [1, 4, 2, 5, 7]

plt.figure(figsize=(4, 4))
plt.plot(x, y, 'o', label='Data Points')
plt.title('Data Set')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

##Linear Interpolation

Linear interpolation presumes that the estimated value falls on the straight line connecting the closest points to the left and right. The corresponding equation is: y(x) = yi + [(yi+1 - yi)(x - xi)] / [xi+1 - xi]. Use linear interpolation to estimate y at x =5 for the data: x=[1,3,6,8] and y=[2,3.5,6,7.5]. Plot the original data and mark the interpolated point.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [1, 3, 6, 8]
y = [2, 3.5, 6, 7.5]

# Create a linear interpolation function
f = interp1d(x, y)
y_hat = f(5)

# Plot
plt.figure(figsize=(6, 6))
plt.plot(x, y, '-ob', label='Linear Interpolation Line')
plt.plot(5, y_hat, 'ro', label=f'Interpolated Point (5, {y_hat:.2f})')
plt.title('Linear Interpolation Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

##Cubic Spline Interpolation

Unlike linear regression, which uses a single function, cubic spline interpolation employs a series of piecewise cubic functions. For a dataset with n points, n-1 cubic functions are required, with each function defined by four coefficients. Use cubic spline interpolation to estimate values between x=0 and
x=6 for the data: x=[0,1,2,3,5,6] and y=[2,3,1,4,2,5]. Plot the original data and the cubic spline interpolation curve.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 5, 6]
y = [2, 3, 1, 4, 2, 5]

# Create cubic spline interpolation
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 6, 200)
y_new = f(x_new)

# Plot
plt.figure(figsize=(8, 8))
plt.plot(x_new, y_new, 'b', label='Cubic Spline Interpolation')
plt.plot(x, y, 'ro', label='Data Points')
plt.title('Cubic Spline Interpolation Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
