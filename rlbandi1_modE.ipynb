##Optimization with Gradient Descent

Use gradient descent to minimize
f(x)=(x‚àí3) ^2.
The gradient is given by
f‚Ä≤(x)=2(x‚àí3).

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return (x - 3)**2

def grad_f(x):
    return 2 * (x - 3)

# Gradient descent parameters
x_current = 0.0      # initial guess
learning_rate = 0.1
num_iterations = 30

# Store values for plotting
x_history = [x_current]

# Gradient descent loop
for i in range(num_iterations):
    x_current = x_current - learning_rate * grad_f(x_current)
    x_history.append(x_current)

print("Estimated minimizer for f(x) = (x-3)^2 is:", x_current)
print("Minimum function value is:", f(x_current))

# Plot the convergence of x
plt.figure(figsize=(8,4))
plt.plot(x_history, marker='o', label='x value')
plt.axhline(y=3, color='r', linestyle='--', label='True minimum (x=3)')
plt.xlabel('Iteration')
plt.ylabel('x value')
plt.title('Gradient Descent Convergence for f(x) = (x-3)^2')
plt.legend()
plt.grid(True)
plt.show()
```

Use gradient descent to minimize the function
ùëì(ùë•,ùë¶)=(ùë•‚àí2)^2+(ùë¶+1)^2.
The gradient is

‚àá
ùëì(ùë•,ùë¶)=

[2(ùë•‚àí2)    
2(ùë¶+1)].


```python
# Define the bivariate function and its gradient
def f_xy(xy):
    x, y = xy
    return (x - 2)**2 + (y + 1)**2

def grad_f_xy(xy):
    x, y = xy
    return np.array([2*(x - 2), 2*(y + 1)])

# Gradient descent parameters for bivariate function
xy_current = np.array([0.0, 0.0])  # initial guess
learning_rate = 0.1
num_iterations = 30

# Store history for plotting
xy_history = [xy_current.copy()]

# Gradient descent loop
for i in range(num_iterations):
    grad = grad_f_xy(xy_current)
    xy_current = xy_current - learning_rate * grad
    xy_history.append(xy_current.copy())

print("Estimated minimizer for f(x,y) is:", xy_current)
print("Minimum function value is:", f_xy(xy_current))

# Convert history to a NumPy array for plotting
xy_history = np.array(xy_history)

# Plot the descent path
plt.figure(figsize=(8,6))
plt.plot(xy_history[:,0], xy_history[:,1], 'o-', label='Descent Path')
plt.plot(2, -1, 'r*', markersize=15, label='True Minimum (2, -1)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Convergence for f(x,y) = (x-2)^2 + (y+1)^2')
plt.legend()
plt.grid(True)
plt.show()
```
